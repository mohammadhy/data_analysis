## Video Game :

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
main_data = pd.read_csv('vgsales.csv')
main_data
```

## as you can see we have Rank columns that not usefull for us so delet this column
## in column Year we have 200 record that has no value 
## task1: cleaning data

```python
main_data.info()
```

```python
main_data = main_data.drop('Rank', axis='columns')
```

```python
main_data['Publisher'].fillna(method='ffill', inplace=True)
main_data['Year'].fillna(method='ffill', inplace=True)
```

```python
main_data.info()
```

## unique value for each columns:

```python
print(main_data['Platform'].unique())
print(main_data['Genre'].unique())
print(main_data['Publisher'].unique())
```

```python
from datetime import datetime
main_data['Year'] = main_data['Year'].astype('int')
main_data['Global_Sales'] = [x * 100 for x in main_data['Global_Sales']]
main_data['Global_Sales'] = main_data['Global_Sales'].astype('int')
main_data['Year'].min()
Q1 = np.quantile(main_data['Year'], 0.25)
Q2 = np.quantile(main_data['Year'], 0.50)
Q3 = np.quantile(main_data['Year'], 0.75)
```

```python
pivote_global = main_data.pivot_table(index='Year', columns='Genre', values='Global_Sales', fill_value=0)
```

```python
pivote_global
```

```python
plt.title('Shooter pllot and action plot ')
plt.plot(pivote_global['Shooter'], label='shooter')
plt.plot(pivote_global['Fighting'],'r--', label='action')
plt.legend()
```

```python
pivote_global.plot.bar()
```

```python
pivote_global.boxplot(vert=False)
```
