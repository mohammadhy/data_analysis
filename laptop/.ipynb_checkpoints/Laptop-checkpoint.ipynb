```python
def clean_model(x):
     return x.split('(')[0]
data['Model Name'] = data['Model Name'].apply(clean_model)
```

```python
def clean_screen(x):
     return x.split(' ')[-1]
data['Resolution'] = data['Screen'].apply(clean_screen)
```

```python
def clean_clock(x):
     return x.split(' ')[-1]
data['clock'] = data['CPU'].apply(clean_clock)
```

```python
def clean_company_cpu(x):
     return x.split(' ')[0]
data['company_cpu'] = data['CPU'].apply(clean_company_cpu)
```

```python
def remove_clock(x):
    clock = x.split(' ')[-1]
    return x.replace(clock, ' ')
data['CPU'] = data['CPU'].apply(remove_clock)
```

```python
def remove_company(x):
    company = x.split(' ')[0]
    return x.replace(company, ' ')
data['model_cpu'] = data['CPU'].apply(remove_company)
```

```python
data['Operating System Version'] = data['Operating System Version'].fillna(0)
```

```python
def os_version(x):
    if x == 0:
        return 'not mention'
    elif x == 10: 
        return x
    elif x == 'X':
        return '10'
    else:
        return x
```

```python
data['Operating System Version'] = data['Operating System Version'].apply(os_version)
```

```python
data['Operating System Version'].value_counts()
```

```python
def os(x):
    if x == 'Windows':
        return x
    elif x == 'No OS': 
        return x
    elif x == 'Linux':
        return x
    elif x == 'Chrome OS':
        return x
    elif x == 'macOS':
        return 'Mac OS'
    else:
        return x
data['Operating System'] = data['Operating System'].apply(os)
```
