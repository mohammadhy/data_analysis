```python
import pandas as pd
import numpy as np
import re
```

```python
#Read Data
data = pd.read_excel('news.xlsx')
```

```python
data.info()
```

```python
data['Date'][10]
```

```python
data['Date'] = data['Date'].apply(lambda x: re.sub('تاریخ انتشار:\n', '', x))
```

```python
data
```

```python
data['month'] = data['Date'].apply(lambda x: x.split()[1])
```

```python
data['month'].value_counts()
```

```python
def convert_to_num(month):
    if month == 'فروردین':
        return 1
    elif month == 'اردیبهشت':
        return 2
    elif month == 'خرداد':
        return 3
    elif month == 'تیر':
        return 4
    elif month == 'بهمن':
        return 11
    elif month == 'اسفند':
        return 12
    else:
        return 'unkhonwn'
```

```python
data['month'] = data['month'].apply(convert_to_num)
```

```python
data
```

```python
data['day'] = data['Date'].apply(lambda x: x.split()[0])
data['year'] = data['Date'].apply(lambda x: x.split()[2])
```

```python
from persiantools import characters, digits
```

```python
data['day'] = data['day'].apply(lambda x: digits.fa_to_en(x))
data['year'] = data['year'].apply(lambda x: digits.fa_to_en(x))
```

```python
data.drop(['Date', 'Title', 'Tag', 'Summary'], axis=1, inplace=True)
```

```python
stopword = []
file = open('stopwords.txt', encoding='UTF-8').read()
[stopword.append(x) for x in file.split()]
stopword
```

```python
def remove_stopwords(text):
    text = str(text)
    filtered_tokens = [token for token in text.split() if token not in stopword]
    filtered_text = ' '.join(filtered_tokens)
    return filtered_text
```

```python
def delete_english_word(text):
    return re.sub(r"['a-zA-Z']", '', str(text))
```

```python
data['Body'] = data['Body'].apply(remove_stopwords)
```

```python
def delete_number(text):
    return re.sub(r"\d+", '', str(text))
```

```python
data['Body'] = data['Body'].apply(delete_number)
```

```python
data['Body'] = data['Body'].apply(delete_english_word)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer, CountVectorizer
```

```python
from sklearn.pipeline import Pipeline
```

```python
model = Pipeline([('vect', CountVectorizer(min_df=40)),('tfidf', TfidfTransformer())])
```

```python
xx = model.fit_transform(data['Body'])
```

```python
df = pd.DataFrame(xx.toarray(), columns=model.named_steps.vect.get_feature_names())
```

```python
df
```

```python
# Creat Model
from sklearn.cluster import KMeans
```

```python
clu = KMeans(n_clusters=3 ,init='k-means++', n_jobs=-1)
```

```python
clu.fit(df)
```

```python
np.unique(clu.labels_, return_counts=True)
```

```python
data['predict_label'] = clu.labels_
```

```python
data['cat'].value_counts()
```

```python
data['predict_label'].value_counts()
```
